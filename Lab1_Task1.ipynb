```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

# Generate grid data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

# 3D plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_title('Z = X² + Y²')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load Pokémon dataset (use your path if needed)
# pokemon = sns.load_dataset("Pokemon.csv").dropna(subset=["hp", "attack", "defense", "sp_attack", "sp_defense", "speed"])
pokemon = df.dropna(subset=["HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"])
print(pokemon.head())

# Select relevant columns
features = pokemon[["HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]]

# Pearson Correlation
pearson_corr = features.corr(method='pearson')

# Spearman Correlation
spearman_corr = features.corr(method='spearman')

# Visualize Pearson
plt.figure(figsize=(8, 6))
sns.heatmap(pearson_corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Pearson Correlation Heatmap")
plt.show()

# Visualize Spearman
plt.figure(figsize=(8, 6))
sns.heatmap(spearman_corr, annot=True, cmap="viridis", fmt=".2f")
plt.title("Spearman Correlation Heatmap")
plt.show()
```
