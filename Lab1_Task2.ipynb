```python
# Step 1: Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Load dataset
df = pd.read_csv('Au_nanoparticle_dataset.csv')

# Step 3: Create new_df with only required columns
new_df = df[["N_total", "N_bulk", "N_surface", "R_avg"]]

# Step 4: Display first 20 samples
print("First 20 samples:")
print(new_df.head(20))

# Step 5: Compute statistics: mean, std, and quartiles
print("\nStatistical Summary:")
print(new_df.describe().T[["mean", "std", "25%", "50%", "75%"]])

# Step 6: Plot histograms in a 1x4 layout
fig, axes = plt.subplots(1, 4, figsize=(20, 4))
features = new_df.columns

for i, col in enumerate(features):
    axes[i].hist(new_df[col], bins=30, color='lightcoral', edgecolor='black')
    axes[i].set_title(f'Histogram of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel("Frequency")

plt.tight_layout()
plt.show()

# Step 7: Pairplot - scatter and histograms
sns.pairplot(new_df)
plt.suptitle("Seaborn Pairplot of Selected Features", y=1.02)
plt.show()

# Step 8: Advanced PairGrid with KDEs and Histograms
g = sns.PairGrid(new_df)

# Bivariate histogram on upper triangle
g.map_upper(sns.histplot)

# Histogram + KDE on diagonal
g.map_diag(sns.histplot, kde=True)

# Bivariate KDE on lower triangle
g.map_lower(sns.kdeplot, fill=True)

plt.suptitle("PairGrid with KDE and Histogram Plots", y=1.02)
plt.show()
```
